```python
import cv2
from google.colab.patches import cv2_imshow

path = "/content/download.jpg"
img = cv2.imread(path)

# line
cv2.line(img,(0,0),(100,200),(10,125,4880))

# Rectangle
cv2.rectangle(img ,(0,0),(100,100),(10,255,255))

# circle
cv2.circle(img,(100,100),58,(0,255,0))

# Text
cv2.putText(img,"My Name is Mudassir",(100,100),cv2.FONT_HERSHEY_COMPLEX_SMALL,1.5,(255,255,0),10)

cv2_imshow(img)
```
