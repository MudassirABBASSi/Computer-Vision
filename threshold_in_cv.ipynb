# threshold is use to find out the object from the pic we convert the rea imag in to gray color and then thresold will apply



```python
import cv2
from google.colab.patches import cv2_imshow
img = "/content/download.jpg"




i = cv2.imread(img)
g = cv2.cvtColor(i,cv2.COLOR_BGR2GRAY)
cv2_imshow(g)
# cv2_imshow(i)
```

#Binary

```python
# Apply a simple binary threshold
# Here, we're setting a threshold of 100. Pixels with intensity > 100 become 255 (white), others become 0 (black).
ret, thresh1 = cv2.threshold(g, 100, 255, cv2.THRESH_BINARY)
thresh1 = cv2.bilateralFilter(thresh1,5,75,75)

ret, thresh1 = cv2.threshold(g, 120, 255, cv2.THRESH_BINARY)
# Display the thresholded image
cv2_imshow(thresh1)
```

```python
im = "/content/images.jpg"
im = cv2.imread(im)

gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(gray,120,255,cv2.THRESH_BINARY_INV)
# ret, thresh2 = cv2.threshold(gray,120,255,cv2.THRESH_BINARY)

print("binary inv")
cv2_imshow(thresh)

thresh6 =cv2.blur(thresh,(75,75))
ret, thresh = cv2.threshold(gray,120,255,cv2.THRESH_BINARY_INV)
print("binary inv1")
cv2_imshow(thresh6)

# print("binary ")
# cv2_imshow(thresh2)





```
