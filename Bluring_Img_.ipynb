```python
import cv2
from google.colab.patches import cv2_imshow
path = "/content/images.jpg"
img = cv2.imread(path)
cv2_imshow(img)

b =cv2.blur(img,(1,5))
cv2_imshow(b)

g = cv2.GaussianBlur(img,(1,5),0)
print("gaussion image")
cv2_imshow(g)

me = cv2.medianBlur(img,5)
print("median")
cv2_imshow(me)
```

#Noise to unnoise image

```python
import cv2
from google.colab.patches import cv2_imshow
path = "/content/download.jpg"
img = cv2.imread(path)
cv2_imshow(img)

b =cv2.blur(img,(1,5))
cv2_imshow(b)

g = cv2.GaussianBlur(img,(1,5),0)
print("gaussion image")
cv2_imshow(g)

me = cv2.medianBlur(img,5)
print("median")
cv2_imshow(me)

a=cv2.bilateralFilter(img,7,75,75)
print("bilateral")
cv2_imshow(a)
```
